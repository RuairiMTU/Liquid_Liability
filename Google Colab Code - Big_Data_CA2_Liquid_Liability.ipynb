My crime datastet has values from 2003 - 2022 but my alcohol licencing dataset only contins information from 2021-2022. I am only going to look at crime data from these years as I cannot make assumptions about the number of alcohol outlets in the preceding years which open my analysis up to mistakes. 

```python
'Filter crime dataset to include values from 2021 and 2022'
df_crime = df_crime[df_crime['Year'].isin([2021, 2022])]
```

I can also see that the Crime data has split the crime locations into the garda station where the crime was committed, which includes information about what Garda division that station is in. Meanwhile, the alcohol data gives outlet locations as an address. I want to categorize the data by garda division and explore the relationship between crime rates in the garda division and the amount of alcohol outlets within that division. 

```python
#create a new colum for the crime data just containing garda division.
#split the garda station colum at the ","
df_crime['Garda Division'] = df_crime['Garda Station'].str.split(',', n=1, expand=True)[1].str.strip()
```

```python
'look at any crime rows that have a null value for the division column to make sure that splitting the column worked correctly'
null_rows = df_crime[df_crime['Garda Division'].isnull()]
display(null_rows)
```

# **Data Trasformation, Analysis & Visualization**

```python
#have an initital look at the alcohol and outlet density counts
#sum the number of crimes in each garda division + make it a dataframe
crime_counts = df_crime.groupby('Garda Division')['VALUE'].sum().reset_index()
#rename value to total crime counts
crime_counts = crime_counts.rename(columns={'VALUE': 'Total Crime Count'})


#count the number of alcohol outlets in each divisions + make it a dataframe
#alcohol_counts = df_alcohol['Garda Division'].value_counts().to_frame()
alcohol_counts = df_alcohol.groupby('Garda Division')['Licence Ref'].count().reset_index()
alcohol_counts = alcohol_counts.rename(columns={'Licence Ref': 'Total Alcohol Outlet Count'})


#look at the data
display(crime_counts)
display(alcohol_counts)

```

```python
#display the descriptive stats in one table
merged_df = pd.merge(crime_counts, alcohol_counts, on='Garda Division')
merged_df = pd.merge(merged_df, df_population_1, on='Garda Division')
#merged_df = merged_df.sort_values(by = "Total Alcohol Outlet Count", ascending = False)
display(merged_df)
```

```python
#display the numbers in a bar graph
# set data from merged_df
x = merged_df['Garda Division']
y1 = merged_df['Total Crime Count']
y2 = merged_df['Total Alcohol Outlet Count']
y3 = merged_df['Population']

# create a figure with three subplots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# plot the crime subplot
ax1.bar(x, y1)
ax1.set_xlabel('Garda Divisions')
ax1.set_ylabel('Count')
ax1.set_title('Crime Count by Garda Division')

# plot the alcohol subplot
ax2.bar(x, y2)
ax2.set_xlabel('Garda Divisions')
ax2.set_ylabel('Count')
ax2.set_title('Alcohol Outlet Number by Garda Division')

#plot the pop subplot
ax3.bar(x, y3)
ax3.set_xlabel('Garda Divisions')
ax3.set_ylabel('Population (Million)')
ax3.set_title('Population Number by Garda Division')

#rotate division titles
ax1.tick_params(axis='x', labelrotation = 90)
ax2.tick_params(axis='x', labelrotation = 90)
ax3.tick_params(axis='x', labelrotation = 90)

# adjust the layout and show the plot
plt.tight_layout()
plt.show()
```

Can see that the as expected the highest populations have the highest amounts of crime and alcohol outlets. Need to normalize data to gather more informative insights - refine the data by normlizing the crime and alcohol counts by accounting for the population of each division. 

```python
#add columns looking at ratios
# create new columns for ratios
merged_df['Normalized Crime'] = merged_df['Total Crime Count'] / merged_df['Population']
merged_df['Normalized Alcohol Outlet Density'] = merged_df['Total Alcohol Outlet Count'] / merged_df['Population']
merged_df['Normalized Crimes Per Alcohol Outlet'] = merged_df['Total Crime Count'] / merged_df['Total Alcohol Outlet Count']
#merged_df = merged_df.sort_values(by = "Normalized Alcohol Outlet Density", ascending = False)

# display the ratios as a table
display(merged_df)
```

Besides outliers and linearity, must meet assumptions of normality and homoscedascity to run a correlation co-efficient, to see if there is any significant relationship between alcohol outlet density and crime rates in the divisions. 

```python
#remove outliers
merged_df_no_outliers = merged_df[(merged_df['Normalized Alcohol Outlet Density'] <= 0.0045) & (merged_df['Normalized Crime'] <= 0.11)]
```

```python


#check for nomality
x = merged_df_no_outliers['Normalized Alcohol Outlet Density']
y = merged_df_no_outliers['Normalized Crime']

# x = merged_df['Normalized Alcohol Outlet Density']
# y = merged_df['Normalized Crime']

# plot histograms to check for normality assumption
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(x, ax=ax[0])
sns.histplot(y, ax=ax[1])
ax[0].set_title('Histogram of Normalized Alcohol Outlet Density')
ax[1].set_title('Histogram of Normalized Crime')

# plot qqplots to check for normality assumption
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
qqplot(x, line='s', ax=ax[0])
qqplot(y, line='s', ax=ax[1])
ax[0].set_title('QQ Plot of Normalized Alcohol Outlet Density')
ax[1].set_title('QQ Plot of Normalized Crime')

plt.show()
```

Alcohol Outlet density is normally distributed but crime rates are not. This may affect future analysis. 

```python
#check for homoscedascity
model = sm.OLS(y, sm.add_constant(x)).fit()


# calculate residuals
residuals = model.resid

# plot scatter plot of residuals against predicted values
sns.residplot(x=model.predict(), y=residuals, lowess=True, scatter_kws={'alpha': 0.5})
plt.xlabel('Predicted Normalized Crime')
plt.ylabel('Residuals')
plt.title('Scatter plot of residuals')
plt.show()
```

The shappe of this graph indicates that the  variances of the two variables are not equal across all levels of the variable. This will affect the validity of any statistical tests I run. 

```python
corr_coef = merged_df_no_outliers['Normalized Alcohol Outlet Density'].corr(merged_df_no_outliers['Normalized Crime'])
#corr_coef = merged_df['Normalized Alcohol Outlet Density'].corr(merged_df['Normalized Crime'])
print(corr_coef)
```

This correlation coefficient suggests a weak negative correlation between the  crime rate and alcohol outlet density in each division
